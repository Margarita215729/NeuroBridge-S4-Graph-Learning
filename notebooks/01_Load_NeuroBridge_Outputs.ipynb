# Phase 2 — Load NeuroBridge-S4 Outputs

**Goal:** Verify that all six processed CSV files from the NeuroBridge-S4 methodology  
can be loaded, validated, and are ready for graph construction in Phase 3.

This notebook is the handoff point between:  
- **NeuroBridge-S4** (biological adaptation scoring)  
- **NeuroBridge-S4-Graph-Learning** (graph representation & ML)

## 0. Setup

```python
import sys
from pathlib import Path
import pandas as pd

repo_root = Path().resolve().parent
if str(repo_root / 'src') not in sys.path:
    sys.path.insert(0, str(repo_root / 'src'))

from neurobridge_graph.data_loader import load_all
print("data_loader imported successfully")
```

## 1. Load all files

```python
data = load_all()

print("Loaded datasets:")
for name, df in data.items():
    print(f"  ✓  {name:<22}  shape={df.shape}")
```

## 2. Pseudo-crew — raw biomarker values

```python
pseudo_crew = data['pseudo_crew']
print(f"Crew members: {len(pseudo_crew)}")
pseudo_crew
```

## 3. Deviation scores — per-biomarker z-scores

```python
dev = data['deviation_scores']
print(f"Shape: {dev.shape}  (participants × biomarkers)")
dev.round(3)
```

## 4. Domain scores — biological domain aggregates

```python
dom = data['domain_scores']
print(f"Domains: {list(dom.columns)}")
dom.round(3)
```

## 5. BACI scores

```python
baci = data['baci_scores']
baci
```

## 6. BACI sensitivity

```python
sens = data['baci_sensitivity']
sens
```

## 7. Crew-level summary

```python
summary = data['crew_level_summary']
summary.T  # transposed for readability
```

## 8. Quick visual — domain scores heatmap

```python
import matplotlib.pyplot as plt
import matplotlib

fig, ax = plt.subplots(figsize=(10, 2.5))
im = ax.imshow(dom.values, cmap='YlOrRd', aspect='auto')
ax.set_xticks(range(len(dom.columns)))
ax.set_xticklabels(dom.columns, rotation=30, ha='right', fontsize=9)
ax.set_yticks(range(len(dom.index)))
ax.set_yticklabels(dom.index, fontsize=9)
ax.set_title('Domain Scores (mean |z-score| per biological domain)', pad=12)
plt.colorbar(im, ax=ax, label='mean |z|')
plt.tight_layout()
plt.show()
```
