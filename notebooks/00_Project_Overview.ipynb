## What is this project?

This repository takes the **processed outputs** of the  
[NeuroBridge-S4 methodology](https://github.com/Margarita215729/NASA-Artemis-II-Human-Research-Data-Methodology-Challenge)  
and builds a **graph-learning layer** on top of them.

The original project produced individual-level biological adaptation scores  
for a proxy Artemis II crew using NHANES reference data.  
This project answers the next question:

> *How are the biological systems of each crew member related to each other,  
> and can we represent that as a graph to enable downstream ML?*

---

## Pipeline overview

```
NeuroBridge-S4 outputs          This repo
(data/processed/)          ──►  Phase 2: Load & validate
                           ──►  Phase 3: Build subject graphs
                           ──►  Phase 4: Extract features
                           ──►  Phase 5: Embeddings & similarity maps
```

## Data files (Phase 2)

| File | Description |
|------|-------------|
| `pseudo_crew.csv` | Raw proxy biomarker values (NHANES subjects) |
| `deviation_scores.csv` | Per-biomarker z-score deviations from reference |
| `domain_scores.csv` | Mean absolute deviation by biological domain |
| `baci_scores.csv` | BACI score + coherence category per crew member |
| `baci_sensitivity.csv` | BACI under varying threshold parameters |
| `crew_level_summary.csv` | Full narrative summary + monitoring priorities |
